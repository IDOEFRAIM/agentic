```python
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import OllamaEmbeddings
from langchain_core.documents import Document
from typing import List, Optional
import logging

logger = logging.getLogger(__name__)

def chroma_vector_database(nodes: List[Document]) -> Optional[Chroma]:
    try:
        embedding = OllamaEmbeddings(model="llama3")
        docsearch = Chroma.from_documents(documents=nodes, embedding=embedding)
        logger.info("Vector database created successfully")
        return docsearch
    except Exception as e:
        logger.error(f"Error in chroma_vector_database: {e}")
        return None

def verify_embeddings(vector_store) -> bool:
    try:
        if hasattr(vector_store, "_collection"):
            count = vector_store._collection.count()
        elif hasattr(vector_store, "index"):
            count = vector_store.index.ntotal
        else:
            logger.warning("Unknown vector store type.")
            return False

        if count == 0:
            logger.warning("No embeddings found in vector store.")
            return False
        else:
            logger.info(f"{count} embeddings found in vector store.")
            return True
    except Exception as e:
        logger.error(f"Error in verify_embeddings: {e}")
        return False
```
