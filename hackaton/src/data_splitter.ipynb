```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_core.documents import Document
from typing import Dict, Any, List
import json, logging

logger = logging.getLogger(__name__)

def split_corpus_data(corpus_data: Dict[str, Any]) -> List[Document]:
    try:
        raw_text = json.dumps(corpus_data, ensure_ascii=False, indent=2)
        document = Document(page_content=raw_text)
        splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
        nodes = splitter.split_documents([document])
        logger.info(f"Created {len(nodes)} nodes from corpus data")
        return nodes
    except Exception as e:
        logger.error(f"Error in split_corpus_data: {e}")
        return []
```
