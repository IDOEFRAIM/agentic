## Creating a Marketing agent with CrewAi :-)

```python
from crewai import Crew,Agent,Task
from langchain_ollama import ChatOllama
```

```python
from crewai import LLM
```

## Setting LLM(model)
We re using a local LLM

```python
llm = LLM(
    model="ollama/mistral",  
    is_litellm=True,
    base_url="http://localhost:11434"
)
```

## Marketing  Agent

```python
marketing_agent = Agent(
    role="Informaticien",
    goal="Design and refine high-impact marketing strategies tailored to audience behavior and platform trends",
    backstory=(
        "You specialize in crafting data-driven marketing strategies that balance creativity with conversion." 
        "You analyze audience signals, platform algorithms, and competitor moves to uncover strategic opportunities." 
        "You prioritize clarity, scalability, and ROI — helping solo creators and teams make smart, actionable decisions."
    ),
    verbose=True,
    llm=llm
)
```

## Task

```python
task = Task(
    description=(
        "Generate a 7-day marketing strategy plan for a product launch targeting Gen Z on TikTok and Instagram." 
        "Include daily focus (e.g. teaser, engagement, conversion), content format, CTA, and platform-specific tactics." 
        "Prioritize virality triggers, audience resonance, and minimal production effort."
    ),
    expected_output='''
    {
        "strategy_plan": [
            {
                "day": "Monday",
                "focus": "Tease product with curiosity-driven reel",
                "platform": "Instagram",
                "format": "Reel (15s), trending audio",
                "cta": "Guess what it is in comments",
                "notes": "Use bold captions and emojis"
            },
            ...
        ]
    }
    ''',
    agent=marketing_agent
)
```

```python
marketing_expert = Crew(
    agents=[marketing_agent],
    tasks=[task]
)
result = crewmar.kickoff()
```
